# Lag plots

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.plotting.lag_plot.html

https://machinelearningmastery.com/time-series-data-visualization-with-python/


A plot of lag 1 is a plot of the values of $Y_i$ versus $Y_{i-1}$
* Vertical axis: $Y_i$ for all $i$
* Horizontal axis: $Y_{i-1}$ for all $i$

Lag plots can provide answers to the following questions:
* Are the data random?
* Is there serial correlation in the data?
* What is a suitable model for the data?
* Are there outliers in the data?

<a href="https://www.itl.nist.gov/div898/handbook/eda/section3/lagplot.htm">source</a>

```python
import pandas
print('pandas',pandas.__version__)
import numpy
print('numpy',numpy.__version__)
import matplotlib.pyplot as plt
```

# linear data

Generate fake data

```python
t = numpy.arange(start=2, stop=3, step=0.1)
y = 5*t+3
_=plt.scatter(t, y)
_=plt.ylabel('y')
_=plt.xlabel('time')
```

### A lag plot is the difference between adjacent values in a series

```python
pandas.plotting.lag_plot(pandas.Series(t));
```

# sine curve

```python
number_of_cycles = 3
number_of_measurements = 500
vshift=3

t = numpy.linspace(0,number_of_cycles*2*numpy.pi,number_of_measurements)

y=numpy.sin(t)+numpy.sin(10*t)+vshift
_=plt.scatter(t,y,s=3)
```

### Lag plot
points cluster along a diagonal line from the bottom-left to the top-right of the plot suggests a positive correlation. (<a href="https://machinelearningmastery.com/time-series-data-visualization-with-python/">source</a>)

```python
pandas.plotting.lag_plot(pandas.Series(y));
```

# Gaussian distribution

A random set of points

```python
num_points=500
data = numpy.random.normal(0,10,num_points)
vshift=5
_=plt.scatter(range(len(data)),data+vshift)
```

### histogram of points

Even though the points are random, they conform to a distribution

```python
bin_count=20
_=plt.hist(data,bins=bin_count, alpha=0.6, color='g')
```

### Lag plot

For random data, the lag plot has no visually obvious correlations

```python
pandas.plotting.lag_plot(pandas.Series(data));
```
