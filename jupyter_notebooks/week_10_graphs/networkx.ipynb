# NetworkX comes with many pre-defined graphs

## binomial graph

The `G_{n,p}` model chooses each of the possible edges with probability p

```python
G = nx.gnp_random_graph(10, 0.2)
nx.draw(G, with_labels=True)
plt.show()
```

```python
A=nx.to_numpy_matrix(G)
print(A.shape)
```

```python
A
```

The square of an adjacency matrix $A^2=(s_{ij})$ has the property that $s_{ij}$ represents the number of walks of length two from vertex $i$ to vertex $j$. [[source]](https://arxiv.org/abs/1207.3122)


If there is an edge $ij$ and an edge $jk$, then there will be a path $ik$ through $j$. The entries $ii$ are the degrees of the nodes $i$. [[source]](https://1stprinciples.wordpress.com/2008/03/30/some-interesting-properties-of-adjacency-matrices/)

```python
# https://networkx.github.io/documentation/stable/reference/algorithms/generated/networkx.algorithms.shortest_paths.unweighted.all_pairs_shortest_path_length.html#networkx.algorithms.shortest_paths.unweighted.all_pairs_shortest_path_length
lengths_dict = dict(nx.all_pairs_shortest_path_length(G))
```

```python
max_length_list=[]
for src_node,dest_and_length in lengths_dict.items():
#    print('source:',src_node)
#    print('destination',dest_and_length)    
#    print(max(dest_and_length.values()))
    max_length_list.append(max(dest_and_length.values()))
print('maximum path length in this graph is',max(max_length_list))
```
